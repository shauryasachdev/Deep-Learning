```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

```python
def sigmoid_derivative(z):
  return sigmoid(z) * (1 - sigmoid(z))
```

```python
def relu(z):
  return np.maximum(0,z)
```

```python
def relu_derivative(z):
  return np.where(z > 0, 1, 0)
```

```python
# Generate a string of random numbers
z = np.linspace(-10, 10, 400)

# Assign these values to the derivatives of sigmoid and ReLu functions

sigmoid_grad = sigmoid_derivative(z)
relu_grad = relu_derivative(z)
```

```python
# Plotting these functions

plt.figure(figsize=(12, 6))

# Plot sigmoid and its derivative
plt.subplot(1,2,1)
plt.plot(z, sigmoid(z), label='Sigmoid Activation', color='b')
plt.plot(z, sigmoid_grad, label='Sigmoid Derivative', color='r', linestyle='--')
plt.title('Sigmoid Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

# Plot sigmoid and its derivative
plt.subplot(1,2,2)
plt.plot(z, relu(z), label='Sigmoid Activation', color='b')
plt.plot(z, relu_grad, label='Sigmoid Derivative', color='r', linestyle='--')
plt.title('ReLu Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()
```
